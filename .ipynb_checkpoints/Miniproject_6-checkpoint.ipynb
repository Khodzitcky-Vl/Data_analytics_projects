```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
# параметры отображения графиков по умолчанию
sns.set(
    font_scale=2,
    style="whitegrid",
    rc={'figure.figsize':(20,7)}
        )
```

```python
# данные взаимодействий с рекламными объявлениями
ads_data = pd.read_csv('../shared/homeworks/python_ds_miniprojects/6/ads_data.csv')
ads_data
```

```python
# Для начала посчитаем среднее количество разных рекламных событий (показы и клики) по объявлениям. 

# Создадим сводную таблицу с кликами и просмотрами для каждого рекламного объявления
click_view_pivot = (ads_data.groupby(['ad_id', 'event'], as_index=False)
 .size()
 .pivot(index= 'ad_id', columns='event', values='size')
 .reset_index()
 .fillna(0.00000000001)
                    )
click_view_pivot
```

```python
# среднее по кликам
click_view_pivot.click.mean()
```

```python
# среднее по объявлениям
click_view_pivot.view.mean()
```

```python
# посмотрим на распределение числа показов объявлений

# для начала прологарифмируем число показов
click_view_pivot['log'] = np.log(click_view_pivot.view)
click_view_pivot
```

```python
sns.distplot(click_view_pivot.log)
```

```python
# Найдем среднее количество показов на 1 объявление (ad_id) по дням
# Для этого воспользуемся сводными таблицами

daily_ad_views = ads_data.loc[ads_data.event == 'view'].pivot_table(
                                                                    index   = 'ad_id',
                                                                    columns = 'date',
                                                                    values  = 'time', 
                                                                    aggfunc = 'count'
                                                                   )
daily_ad_views.head()
```

```python
# посчитаем скользящее среднее по полученному результату с окном в 2 элемента.
daily_ad_views_mean = daily_ad_views.mean()
daily_ad_views_rolling = daily_ad_views_mean.rolling(2).mean()
daily_ad_views_rolling
```

```python
# Альтернативный способ выполнения предыдущей задачи: 

ads_dates = (ads_data.loc[ads_data.event == 'view']
             .groupby(['date', 'ad_id'], as_index=False)
             .size()
            )
ads_dates_2 = (ads_dates.groupby('date')
                        .agg({'size':'mean'})
            )
ads_dates_2.rolling(2).mean()
```

```python
# Нанесем на один график значения просто среднего количества показов по дням 
# и скользящего среднего на основе этих данных

fig, ax = plt.subplots()
sns.lineplot(data=daily_ad_views_mean, label='mean')
sns.lineplot(data=daily_ad_views_rolling, label='rolling_mean')
```

```python
# Посчитаем , в какой день наблюдается наибольшая разница 
# по модулю между арифметическим средним и скользящим средним

# Для удобства соединим данные в один новый датафрейм

mean_merged = pd.DataFrame()
mean_merged['mean'] = daily_ad_views_mean
mean_merged['rolling_mean'] = daily_ad_views_rolling
mean_merged = mean_merged.reset_index().rename(columns={'index':'date'})

mean_merged['difference'] = abs(mean_merged['mean'] - mean_merged['rolling_mean'])
mean_merged.sort_values('difference', ascending = False)
```

```python
# выгрузим данные о характеристиках рекламных клиентов
ads_clients_data = pd.read_csv('../shared/homeworks/python_ds_miniprojects/6/ads_clients_data.csv', parse_dates=['date', 'create_date'])
ads_clients_data
```

```python
# найдем среднее количество дней от даты создания рекламного клиента (create_date) 
# и первым запуском рекламного объявления этим клиентом (date)

ads_data_full = ads_data.merge(ads_clients_data, on='client_union_id', suffixes=('_ad_data', '_client_data' ))
ads_data_full.date_ad_data = ads_data_full.date_ad_data.astype('datetime64[ns]')
ads_data_full.
```

```python
ads_data_full['first_ad_time'] = ads_data_full.date_ad_data - ads_data_full.create_date

first_ad = ads_data_full.groupby('client_union_id', as_index=False).first_ad_time.min()
first_ad
```

```python
first_ad.first_ad_time.mean()
```

```python
# Вычислим конверсию из создания рекламного кабинета в запуск первой рекламы в течение не более 365 дней

threshold = pd.Timedelta(365, unit='day')
before365_days = first_ad.loc[first_ad.first_ad_time <= threshold].client_union_id.count()
(before365_days/ads_clients_data.client_union_id.nunique() * 100).round(2)
```

```python
# Определим, сколько уникальных клиентов запустили свое первое объявление 
# в первый месяц своего существования (от 0 до 30 дней)

# Сначала  разобъем наших клиентов по промежуткам 
# от создания рекламного кабинета до запуска первого рекламного объявления
bins = pd.to_timedelta([0, 30, 90, 180, 365], 'd')
first_ad['cut'] = pd.cut(first_ad.first_ad_time, bins, labels=['0-30', '30-90', '90-180', '180-365'])
first_ad
```

```python
first_ad.cut.value_counts().sort_values()
```
