```python
 #Проверим пользователи из каких стран совершили наибольшее число успешных бронирований. Бронирование считается успешным, если в дальнейшем не было отменено 
bookings.loc[bookings['is_canceled'] == 0]
bookings.value_counts('country') 
#раньше "value_counts" был только методом для Series, теперь его можно использовать и с DataFrame. Но тут на платформе старая версия Pandas
```

```python
# На сколько ночей (stays_total_nights)  в среднем бронируют отели типа City Hotel? Resort Hotel?
(bookings.groupby('hotel') 
                 .stays_total_nights.mean()
                 .round(2)
                )
```

```python
(bookings.query('arrival_date_year == 2016')
 .value_counts('arrival_date_month')
)
# Проверим на какой месяц чаще всего оформляли бронь в 2016 году
```

```python
# А в 2017? 
(bookings.query('arrival_date_year == 2017')
 .value_counts('arrival_date_month')
)
```

```python
# Проверим, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в 2015, 2016 и 2017.
(bookings.query('hotel == "City Hotel" and is_canceled == 1')
         .groupby('arrival_date_year')['arrival_date_month']
         .value_counts()
)
```

```python
# найдем наибольшее среднее количество бронирований среди разных возрастов 
bookings[['adults', 'children', 'babies']].mean().idxmax()
```

```python
# Найдем среднее количество детей, отдыхающих в отелях разных видов
bookings['total_kids'] = bookings['children'] + bookings['babies']
(bookings.groupby('hotel')
         .total_kids.mean()
         .round(2)
)
```
