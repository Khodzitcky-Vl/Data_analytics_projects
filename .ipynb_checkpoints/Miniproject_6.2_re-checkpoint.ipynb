```python
# Надо найти самое частое значение parent
# Попробуем решить эту задачу без использования pandas

path_to_file = '../shared/homeworks/python_ds_miniprojects/6_gff/gff.tsv'
gff_data = open(path_to_file)
gff_data = gff_data.readlines()
gff_data
```

```python
list_of_parents = []
for line in gff_data:
    line = line.split('Parent=')
    list_of_parents.append(line[-1].strip('\n'))
list_of_parents
```

```python
values, counts = np.unique(list_of_parents, return_counts=True)
values, counts
```
